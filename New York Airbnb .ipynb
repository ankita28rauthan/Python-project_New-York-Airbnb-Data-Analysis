Task 4: Data Cleaning

```python
data.isnull().sum()
# dropping all missing rows
data.dropna(inplace=True)
data.isnull().sum()
```

```python
data.shape
```

```python
#dealing with duplicates rows
data.duplicated().sum()
#data[data.duplicated()]
#deleting all duplicated rows
data. drop_duplicates(inplace=True)
data.duplicated().sum()
```

```python
data.dtypes
```

```python
#type casting
# changing data types

data['id']=data['id'].astype(object)
data['host_id']=data['host_id'].astype(object)
data.dtypes
```

Task 5: Data Analysis

Univariate Analysis

```python
# identifying outliers in price

df= data[data['price'] < 1500]

sns.boxplot(data=df,x='price')
```

```python
#Price Distribution
plt.figure(figsize=(8,5))
sns.histplot(data=df,x='price',bins=100)
plt.title('Price Distribution')
plt.ylabel("Frequency")
plt.show()
```

```python
df.columns
```

```python
#Availability Distribution
plt.figure(figsize=(6,3))
sns.histplot(data=df,x='availability_365')
plt.title('availability_365 Distribution')
plt.ylabel("Frequency")
plt.show()
```

```python
df.groupby(by='neighbourhood_group')['price'].mean()
```

```python
#Average Price per bed
df.groupby(by='neighbourhood_group')['price per bed'].mean()
```

***Feature Engineering***

```python
#['price per bed']
df.loc[:,'price per bed'] = df['price'] / df['beds']
df.head()
```

***Bi Varibale Analysis***

One Variable dependency in another variables

```python
df.columns
```

```python
#price dependency on neighbourhood
sns.barplot(data=df,x='neighbourhood_group',y='price',hue='room_type')
```

```python
#number of reviews and price relationship
plt.figure(figsize=(8,5))
plt.title('Locality and Review Dependency')
sns.scatterplot(data=df,x='number_of_reviews',y='price',hue='neighbourhood_group')
```

```python
df.dtypes
```

```python
sns.pairplot(data=df, vars=['price','minimum_nights','number_of_reviews','availability_365'],hue='room_type')
```

```python
sns.countplot(data=df, x='room_type')
plt.title('Number of Listings by Room Type')
plt.ylabel('Count')
plt.show()
```

```python
sns.scatterplot(data=df, x='availability_365', y='price', hue='room_type')
plt.title('Availability vs Price')
plt.show()
```

```python
#Geographical Ditribution of AirBnb Listings
plt.figure(figsize=(10,7))
sns.scatterplot(data=df,x='longitude',y='latitude',hue='room_type')
plt.title("Geographical Distribution Of AirBnb Listing in New York City by Room Type")
plt.show()
```

```python
#heat map- correlation of one variable with others for numerical column

corr = df[['latitude','longitude','price','minimum_nights','number_of_reviews','reviews_per_month','availability_365']].corr()

corr
plt.figure(figsize=(8,6))
sns.heatmap(data=corr,annot=True)
```
